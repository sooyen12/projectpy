```python
import pandas as pd

# 업로드한 파일의 이름을 사용하여 Excel 파일을 읽어옵니다.
one_person_household= pd.read_excel('/content/drive/MyDrive/학교 수업/2025년 1학기 /2025 1학기 - 시계열데이터분석/시계열 프로젝트/1인가구_성별_나이_지역.xlsx')

print(one_person_household)
```

```python
one_person_household_copy['교차변수'] = one_person_household_copy['행정구역별(시군구)'] + '_' + one_person_household_copy['성별'] + '_' + one_person_household_copy['연령'].astype(str)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 연도 컬럼만 추출
year_cols = [str(year) for year in range(2015, 2023)]  # 2015~2022

# 그래프 그리기
plt.figure(figsize=(14, 8))

# 각 교차변수(성별+연령 조합)에 대해 선 그래프 그리기
for i in range(len(filtered_df)):
    label = filtered_df.loc[i, '교차변수']
    values = filtered_df.loc[i, year_cols].astype(int)  # 값들을 숫자로 변환
    plt.plot(year_cols, values, label=label)

plt.title('연령 및 성별별 1인 가구 수 변화 (2015~2022)')
plt.xlabel('연도')
plt.ylabel('가구 수')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # 범례를 그래프 바깥에 배치
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 연도 컬럼 설정
year_cols = [str(year) for year in range(2015, 2023)]

# 전국 남자 필터링
male_df = filtered_df[filtered_df['성별'] == '남자']

# 그래프
plt.figure(figsize=(12, 7))
for i in range(len(male_df)):
    label = male_df.iloc[i]['연령']
    values = male_df.iloc[i][year_cols].astype(int)
    plt.plot(year_cols, values, label=label)

plt.title('전국 남자 연령별 1인 가구 수 변화 (2015~2022)')
plt.xlabel('연도')
plt.ylabel('가구 수')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# 전국 여자 필터링
female_df = filtered_df[filtered_df['성별'] == '여자']

# 그래프
plt.figure(figsize=(12, 7))
for i in range(len(female_df)):
    label = female_df.iloc[i]['연령']
    values = female_df.iloc[i][year_cols].astype(int)
    plt.plot(year_cols, values, label=label)

plt.title('전국 여자 연령별 1인 가구 수 변화 (2015~2022)')
plt.xlabel('연도')
plt.ylabel('가구 수')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# 구글 드라이브에 있는 엑셀 파일 경로
file_path = '/content/drive/MyDrive/학교 수업/2025년 1학기 /2025 1학기 - 시계열데이터분석/시계열 프로젝트/범죄발생.xlsx'  # 파일 경로

# 엑셀 파일 읽기
df = pd.read_excel(file_path)
```

```python
df
```

```python
# 0번째 줄 삭제 (행 삭제)
df = df.drop(0, axis=0)

# A열(0번째 열) 삭제
df = df.drop(df.columns[0], axis=1)

# '죄종별(2)' 열 이름을 '죄종별'로 변경
df = df.rename(columns={'죄종별(2)': '죄종별'})

# 결측값(NaN) 제거: NaN이 포함된 행을 삭제
crime_df = df.dropna()

# 데이터프레임을 표 형식으로 출력
from IPython.display import display
display(crime_df)
```
